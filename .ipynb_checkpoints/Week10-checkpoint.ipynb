```python
##Load dataframe and remove two column
import pandas as pd

data = pd.read_csv('./Files/iris_data.csv', decimal=",")
data = data.drop(['Petal length', 'Petal width'], axis=1)
data
```

```python
#Get unique labesl
unique_labels= data.Species.unique()
unique_labels
```

```python
#plot with a scatter plot each iris flower sample colored by label (3 different colors)
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

colors ={unique_labels[0] : 'blue', unique_labels[1] : 'green', unique_labels[2] : 'red'}

color_list = []
for spec in data.Species:
    color_list.append(colors[spec])
    
color_list
ax.scatter(data['Sepal length'], data['Sepal width'], c=color_list)
```

```python
#use: MeanShift and estimate_bandwidth from sklearn.cluster to first estimate bandwidth and then get the clusters (HINT: estimate_bandwidth() takes an argument: quantile set it to 0.2 for best result)
from sklearn.cluster import MeanShift, estimate_bandwidth
import numpy as np

bandwidth = estimate_bandwidth(data, quantile=0.2)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(data)
labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters = len(labels_unique)
```

```python
#print out labels, cluster centers and number of clusters (as returned from the MeanShift function)
print(labels, cluster_centers, n_clusters)
```

```python
#create a new scatter plot where each flower is colored according to cluster label
#add a dot for the cluster centers

from itertools import cycle
from sklearn.datasets import make_blobs

fig2= plt.figure()
ax2 = fig2.add_subplot(111)
colors = cycle('bgrcmy')
for k, col, in zip(range(n_clusters), colors):
    my_member = (labels == k)
    cluster_center = cluster_centers[k]
    
    x, y = data.to_numpy()[my_member, 0], data.to_numpy()[my_member, 1]
    ax2.scatter(x, y, c=col, linewidth=0.2)
    ax2.scatter(cluster_center[0], cluster_center[1], c='k', s=50, linewidth=0.2)
    
plt.title('Estimated number of clusters: {}'.format(n_clusters))
```
