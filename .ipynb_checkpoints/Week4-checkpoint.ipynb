#### Exercise 1
1. Open the file './data/befkbhalderstatkode.csv'
2. Turn the csv file into a numpy ndarray with `np.genfromtxt(filename, delimiter=',', dtype=np.uint, skip_header=1)`
3. Using this data:
```python
neighb = {1: 'Indre By', 2: 'Østerbro', 3: 'Nørrebro', 4: 'Vesterbro/Kgs. Enghave', 
          5: 'Valby', 6: 'Vanløse', 7: 'Brønshøj-Husum', 8: 'Bispebjerg', 9: 'Amager Øst', 
          10: 'Amager Vest', 99: 'Udenfor'}
```
Find out how many people lived in each of the 11 areas in 2015
4. Make a bar plot to show the size of each city area from the smallest to the largest
5. Create a boolean mask to find out how many people above 65 years lived in Copenhagen in 2015
6. How many of those were from the other nordic countries (not dk). Hint: see notebook: "04 Numpy"
7. Make a line plot showing the changes of number of people in vesterbro and østerbro from 1992 to 2015

````python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)

file_path = '.././data/befkbhalderstatkode.csv'

neighb = {1: 'Indre By', 2: 'Østerbro', 3: 'Nørrebro', 4: 'Vesterbro/Kgs. Enghave', 
       5: 'Valby', 6: 'Vanløse', 7: 'Brønshøj-Husum', 8: 'Bispebjerg', 9: 'Amager Øst', 
       10: 'Amager Vest', 99: 'Udenfor'}

nlist = np.genfromtxt(file_path, delimiter=',', dtype=np.uint, skip_header=1)
dd = nlist

plot_list = {}

for key, value in neighb.items():
    mask = (dd[:,0] == 2015) & (dd[:,1] == key) & (dd[:,3] == 5100)
    people = 0
    for element in dd[mask]:
        people += element[4]
    print('In', value, ' lives: ', people)
    people_list.append(people)
    plot_list[value] = people
    
import matplotlib.pyplot as plt
import operator

plt.bar(range(len(plot_list)), list(plot_list.values()), width=0.5)
plt.xticks(range(len(plot_list)), list(plot_list.keys()), rotation=90)
plt.show()

sorted_list = dict(sorted(plot_list.items(), key=operator.itemgetter(1), reverse=False))
plt.bar(range(len(sorted_list)), list(sorted_list.values()), width=0.5)
plt.xticks(range(len(sorted_list)), list(sorted_list.keys()), rotation=90)
plt.show()

mask_greater65 = (dd[:,0] == 2015) & (dd[:,2] > 65)
people = 0
for line in dd[mask_greater65]:
    people += line[4]
print('There is:', people, 'over 65 in Copenhagen')

nordic = {'Finland' : 5104, 'Norge': 5110, 'Sverige': 5120}

for key, value in nordic.items():
    mask_nordic = (dd[:,0] == 2015) & (dd[:,2] > 65) & (dd[:,3] == value)
    people = 0
    for element in dd[mask_nordic]:
        people += element[4]
    print('There were', people, 'from', key, 'living in copenhagen in 2015')
    

people_vesterbro = []
people_østerbro = []
year = range(1992, 2016)


for x in year:
    mask_vesterbro = (dd[:,0] == x) & (dd[:,1] == 4)
    people = 0
    for element in dd[mask_vesterbro]:
        people += element[4]
    people_vesterbro.append(people)

    
for x in year:
    mask_østerbro = (dd[:,0] == x) & (dd[:,1] == 2)
    people = 0
    for element in dd[mask_østerbro]:
        people += element[4]
    people_østerbro.append(people)
  
plt.plot(year, people_vesterbro, color='red', marker='o')
plt.title('Number Of People in Vesterbro Per Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number Of People', fontsize=14)
plt.grid(True)
plt.show()

plt.plot(year, people_østerbro, color='green', marker='o')
plt.title('Number Of People in Vesterbro Per Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number Of People', fontsize=14)
plt.grid(True)
plt.show()


print(people_vesterbro)
print(people_østerbro)
````
