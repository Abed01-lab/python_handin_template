```python
class Student():
    
    def __init__(self, name, gender, dataSheet, image_url):
        self.name = name
        self.gender = gender
        self.dataSheet = dataSheet
        self.image_url = image_url
        self.avg_grade = self.get_avg_grade
        self.etcs = self.get_etcs
    
    def __str__(self):
        return '{name}, {gender}, {grades}, {avg_grades}'.format(
            name=self.name, gender=self.gender,
            grades=self.dataSheet.get_grades_as_list(),
            avg_grades=self.avg_grade())
    
    def get_avg_grade(self):
        grades = self.dataSheet.get_grades_as_list()
        length = len(grades)
        total = sum(grades)
        return total / length
    
    def get_etcs(self):
        etcs = (self.dataSheet.get_etcs_points() * 100) / 150
        rounded_etcs = round(etcs, 2)
        return rounded_etcs
        
```

```python
class DataSheet():
    
    def __init__(self, courses=[]):
        self.courses = courses
        self.grades = []
        self.etcs = 0
    
    def get_grades_as_list(self):
        self.grades = []
        for course in self.courses:
            self.grades.append(course.optionalGrade)
        return self.grades
    
    def get_etcs_points(self):
        self.etcs = 0
        for course in self.courses:
            self.etcs += course.ETCS
        return self.etcs
```

```python
class Course():
    
    def __init__(self, name, classroom, teacher, ETCS, optionalGrade):
        self.name = name
        self.classroom = classroom
        self.teacher = teacher
        self.ETCS = ETCS
        self.optionalGrade = optionalGrade
    
```

### 8. Read student data into a list of Students from a csv file (Each student can appear on multiple lines): 
   1. loop through the list and print each student with name, img_url and avg_grade.
   2. sort the list by avg_grade
   3. create a bar chart with student_name on x and avg_grade on y-axis

```python
import csv

file_path = './Files/Students.csv'
students = []
names = []
avg_grades = []
with open(file_path, 'r') as f:
    content = f.readlines()
    
    for line in content:
        line_to_append = line.strip().split(',')
        students.append(line_to_append)
        print(line_to_append)

students.pop(0)
for student in students:
    names.append(student[0])
    avg_grades.append(student[3])
    
print(names)
print(avg_grades)
%matplotlib notebook
import matplotlib.pyplot as plt
plt.bar(names, avg_grades)
```
